# Init Embedding Model

```python
import openai
def get_embedding_from_api(text, model="text-embedding-ada-002"):
   text = text.replace("\n", " ")
   return openai.Embedding.create(input = [text], model=model)['data'][0]['embedding']

def print_vector_details(vector):
    print(vector)
    print(len(vector))
```

# Single Word Embeddings

```python
man = get_embedding_from_api("man")
print_vector_details(man)
```

```python
women = get_embedding_from_api("women")
print_vector_details(women)
```

```python
king = get_embedding_from_api("King")
print_vector_details(king)
```

```python
queen = get_embedding_from_api("Queen")
print_vector_details(queen)
```

```python
from openai.embeddings_utils import cosine_similarity

def similarity_v(v1, v2):
    return cosine_similarity(v1,v2)
```

```python
print(similarity_v(man,man))
print(similarity_v(man,king))
print(similarity_v(man,women))
print(similarity_v(man,queen))
print(similarity_v(man,get_embedding_from_api("Sunny beaches in Hawaii are great for surfing.")))
```

# Multi Word Embeddings

```python
from openai.embeddings_utils import get_embedding, cosine_similarity

# Get embeddings for two pieces of text
text1 = "The weather in New York City is rainy and cold today."
text2 = "This is an eggplant"
text3 = "Today, the climate in New York City is characterized by rain and chilly temperatures. "
embedding1 = get_embedding(text1)
embedding2 = get_embedding(text2)
embedding3 = get_embedding(text3)
```

# Check Similarity

```python
# Calculate the cosine similarity between the two vectors
similarity1 = cosine_similarity(embedding1, embedding2)

# Print the similarity score
print("Cosine Similarity:", similarity1)
```

```python
similarity2 = cosine_similarity(embedding1, embedding3)
print("Cosine Similarity:", similarity2)
```
